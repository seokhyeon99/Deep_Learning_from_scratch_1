## 2.3 How to implement perceptron

## 2.3.1 simple implementation

- AND logical circuit implementation

```python
def AND(x1, x2):
    w1, w2, theta = 1, 1, 1.5
    tmp = w1*x1 + w2*x2
    if tmp > theta:
        return 1
    else:
        return 0
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

- OR gate with weights and bias

```python
def OR_bias(x1, x2):
    x = np.array([x1, x2])
    w = np.array([1, 1])
    b = -0.5
    tmp = w * x + b
    if tmp > 0:
        return 1
    else:
        return 0
```
